## 2. Cleaning
HTML, URLs, emojis, hashtags, digits, punctuation removed; lowercase and whitespace normalized.

## 3. Flow
Raw → Cleaned

```python
def clean(t):
 t=re.sub(r'<[^>]+>',' ',t)
 t=re.sub(r'https?://\S+|www\.\S+',' ',t)
 t=re.sub(r'[\U0001F300-\U0001FAFF\U00002700-\U000027BF]',' ',t)
 t=re.sub(r'#\w+',' ',t)
 t=re.sub(r'\d+',' ',t)
 t=re.sub(r'[^\w\s]',' ',t)
 return re.sub(r'\s+',' ',t).strip().lower()
cleaned=[clean(x) for x in corpus]
pd.DataFrame({'Raw':corpus,'Cleaned':cleaned})
```

## Optional preprocessing
Stopword removal, stemming, and lemmatization are shown as optional because modern deep learning often skips them.

```python
import nltk
nltk.download('stopwords',quiet=True); nltk.download('wordnet',quiet=True)
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
sw=set(stopwords.words('english'))
ps=PorterStemmer(); lm=WordNetLemmatizer()
demo=pd.DataFrame({
'Cleaned':cleaned,
'No Stopwords':[' '.join(w for w in s.split() if w not in sw) for s in cleaned],
'Stemmed':[' '.join(ps.stem(w) for w in s.split()) for s in cleaned],
'Lemmatized':[' '.join(lm.lemmatize(w) for w in s.split()) for s in cleaned]})
demo
```

## Tokenization and Vocabulary
Flow: Cleaned Text → Tokens → Integer IDs

```python
tok=Tokenizer(oov_token='<OOV>')
tok.fit_on_texts(cleaned)
seq=tok.texts_to_sequences(cleaned)
pd.DataFrame(tok.word_index.items(),columns=['Word','ID']).head(20)
```

## Integer Sequences

```python
pd.DataFrame({'Sentence':cleaned,'Sequence':seq})
```

## Padding Examples
Compare pre-padding and post-padding.

```python
maxlen=max(map(len,seq))
post=pad_sequences(seq,maxlen=maxlen,padding='post')
pre=pad_sequences(seq,maxlen=maxlen,padding='pre')
print('Max length:',maxlen)
pd.DataFrame({'Original':seq,'Post Padding':post.tolist(),'Pre Padding':pre.tolist()})
```

## Truncation Examples
Illustrates truncating long sequences from the beginning or end.

```python
long_seq=[list(range(1,16))]
post_trunc=pad_sequences(long_seq,maxlen=8,truncating='post')
pre_trunc=pad_sequences(long_seq,maxlen=8,truncating='pre')
print('Original:',long_seq[0])
print('Post truncation:',post_trunc[0].tolist())
print('Pre truncation :',pre_trunc[0].tolist())
```
